```python
import os
from pathlib import Path

def setup_colab_environment():
    """
    Configura a infraestrutura de diretГіrios e arquivos necessГЎria para o ChessIt.
    Complexidade: O(k) onde k Г© o nГєmero de diretГіrios/arquivos.
    """
    print("рџљЂ Iniciando configuraГ§ГЈo do ambiente ChessIt...")

    # 1. DefiniГ§ГЈo da estrutura de pastas
    # 'training/data' para o PGN, 'models' para o output inicial, 'bin' para o final
    directories = [
        'training/data',
        'models',
        'bin'
    ]

    for folder in directories:
        Path(folder).mkdir(parents=True, exist_ok=True)
        print(f"вњ… DiretГіrio garantido: {folder}")

    # 2. CriaГ§ГЈo do arquivo requirements.txt
    # Baseado exatamente no conteГєdo que vocГЄ enviou
    requirements_content = """torch
onnx
onnxruntime
python-chess
tqdm
numpy
"""
    req_path = Path("training/requirements.txt")
    with open(req_path, "w", encoding="utf-8") as f:
        f.write(requirements_content)
    print(f"вњ… Arquivo criado: {req_path}")

    # 3. MovimentaГ§ГЈo lГіgica (Opcional/SeguranГ§a)
    # Se os scripts .py estiverem na raiz, movemos para /training/ para respeitar seus comandos
    scripts = ['train_chessit_model.py', 'inline_onnx_external_data.py']
    for script in scripts:
        if os.path.exists(script):
            os.replace(script, f"training/{script}")
            print(f"рџљљ Script {script} movido para pasta /training/")

    print("\nрџЏЃ Ambiente pronto para instalaГ§ГЈo de pacotes e treinamento.")

# Executa a funГ§ГЈo de setup
setup_colab_environment()
```
